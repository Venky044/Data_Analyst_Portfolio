### Cleaning Columns with Symbols(UoM)
Remove unnecessary characters such as 'cm', 'kg', 'lbs', 'K'

```python
# Examining the Value, Wage and Release_Clause columns together
fifa_df.loc[:,['Height', 'Weight', 'Hits', 'Value', 'Wage', 'Release_Clause', 'W/F', 'SM', 'IR']]
```

#### After Examining the above columns:
* Conversion of cm, inch & ft are needed to the Height column
* Coversion of kg & lbs are needed to the Weight column
* Coversion for Euros(€), Thousand(K) & Million(M) are needed to the Value & Wage column
* there is a * attacted with every values in the ['W/F', 'SM', 'IR] columns are needed to remove the *symble

```python
# Function for converting object value to float

def obj_to_float(value):
    if isinstance(value, str):
        if value.find('cm')!=-1:
            value=value.replace('cm','')
            return float(value)
        elif value.find("'")!=-1:
            value_split=value.split("\'")
            value=(float(value_split[0])*12+float(value_split[1].replace('"','')))*(2.54)
            return value
        elif value.find('kg')!=-1:
            value=value.replace('kg','')
            return float(value)
        elif value.find('lbs')!=-1:
            value=value.replace('lbs','')
            return float(value)*0.453592
        elif value.find('K')!=-1:
            value=value.replace('K','').replace('€','')
            return float(value)*1000
        elif value.find('M')!=-1:
            value=value.replace('M','').replace('€','')
            return float(value)*10e6
        elif value.find('€')!=-1:
            value=value.replace('€','')
            return float(value)
        elif value.find('★')!=-1:
            value=value[:1]
            return int(value)
        else:
            return value
    else:
        return value 
```

```python
# Applying function to Height column
fifa_df['Height(cm)']=fifa_df['Height'].apply(obj_to_float)
fifa_df.drop(['Height'], axis=1, inplace=True)

# Applying function to Weight column
fifa_df['Weight(kg)']=fifa_df['Weight'].apply(obj_to_float)
fifa_df.drop(['Weight'], axis=1, inplace=True)

# Applying function to Value column
fifa_df['Value(€)']=fifa_df['Value'].apply(obj_to_float)
fifa_df.drop(['Value'], axis=1, inplace=True)

# Applying function to Wage column
fifa_df['Wage(€)']=fifa_df['Wage'].apply(obj_to_float)
fifa_df.drop(['Wage'], axis=1, inplace=True)

# Applying function to Release_Clause column
fifa_df['Release_Clause(€)']=fifa_df['Release_Clause'].apply(obj_to_float)
fifa_df.drop(['Release_Clause'], axis=1, inplace=True)

# Applying function to W/F column
fifa_df['W/F']=fifa_df['W/F'].apply(obj_to_float)

# Applying function to SM column
fifa_df['SM']=fifa_df['SM'].apply(obj_to_float)

# Applying function to IR column
fifa_df['IR']=fifa_df['IR'].apply(obj_to_float)

# Applying function to Hits column
fifa_df['Hits']=fifa_df['Hits'].apply(obj_to_float)
```
